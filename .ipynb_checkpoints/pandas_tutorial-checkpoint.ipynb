## Missing values

```python
# NaN values in pandas come from numpy
import numpy as np
np.nan
```

```python
df.isna().head(3)
```

```python
df.isnull().head(3)
```

```python
df.notna().head(3)
```

```python
df.isna().any()
```

```python
df.isna().sum(axis='rows')
```

```python
# Remove missing values from subset. how{‘any’, ‘all’}, default ‘any’, axis{0 or ‘index’, 1 or ‘columns’}, default 0, subset optional.
df_clean = df.dropna(axis='index', how='any', thresh=None, subset=['type2', 'weight_kg', 'percentage_male'], inplace=False)
df_clean.isna().sum()
```

```python
# Fill NA/NaN values using the specified method. ffill “forward fill” replaces missing values with the values in the previous row. bfill “backward fill”.
# axis{0 or ‘index’, 1 or ‘columns’} Axis along which to fill missing values.
df_clean = df.fillna(method='ffill', axis='index', inplace=False)
df_clean.head(3)
```

```python
# fill missing values with mean of the column
df_clean = df.fillna(df.mean(), inplace=False)
df_clean.isna().sum()
```

```python
df_clean = df.fillna({'type2': 10, 'weight_kg': 10, 'percentage_male': 10}, inplace=False)
df_clean.isna().sum()
```

## Element-wise operations: add, sum, mul, div, etc
By default these DataFrame methods operate accross columns.

```python
# Return the sum of the values over the requested axis. This is equivalent to the method numpy.sum.
df['dummy'] = df[['attack', 'defense']].sum(axis='columns')
df.head(3)
```

```python
df['dummy'] = df['attack'] + df['defense']
df.head(3)
```

```python
df['dummy'] = df['attack'] + 12.35
df.head(3)
```

```python
import numpy as np
df['dummy'] = df['attack'] + np.random.randn(df.shape[0])
df.head(3)
```

```python
# Get Multiplication of dataframe and other, element-wise (binary operator mul). Equivalent to dataframe * other, but with support to substitute a fill_value for missing data in one of the inputs.
df['dummy'] = df['attack'].mul(df['defense'], fill_value=1)
df.head(3)
```

```python
df['dummy'] = df['attack'] * df['defense']
df.head(3)
```

```python
# Get Floating division of dataframe and other, element-wise (binary operator truediv). Equivalent to dataframe / other, but with support to substitute a fill_value for missing data in one of the inputs.
df['dummy'] = df['attack'].div(df['defense'], fill_value=1)
df.head(3)
```

```python
df['dummy'] = df['attack'] / df['defense']
df.head(3)
```

```python
df['dummy'] = df['attack'] // df['defense']
df.head(3)
```

```python
df['dummy'] = df['attack'] % df['defense']
df.head(3)
```

```python
df['dummy'] = df['attack'] ** df['defense']
df.head(3)
```

```python
df['dummy'] = df['attack'] - df['defense']
df.head(3)
```

```python
if 'dummy' in df.columns:
    df.drop(columns=['dummy'], inplace=True)
```

## Column wise operations

```python
df.mean(axis='columns')
```

```python
#Return the maximum of the values over the requested axis. If you want the index of the maximum, use idxmax. This isthe equivalent of the numpy.ndarray method argmax.
df.max(axis='index', numeric_only=True)
```

```python
# Count non-NA cells for each column or row. The values None, NaN, NaT, and optionally numpy.inf (depending on pandas.options.mode.use_inf_as_na) are considered NA.
# OBS for binary series and DataFrames it counts both True and False (not NaNs). Use .value_counts() or .sum() instead.
df.count(axis='index')
```

## Group by

```python
df.groupby(['type1', 'type2'], sort=True, group_keys=True)
```

```python
next(iter(df[['type1', 'type2', 'name', 'attack', 'defense']].groupby(['type1', 'type2'])))[0]
```

```python
next(iter(df[['type1', 'type2', 'name', 'attack', 'defense']].groupby(['type1', 'type2'])))[1]
```

```python
df['count'] = 1
df[['type1', 'type2', 'count']].groupby(['type1', 'type2']).count()
```

```python
# Compute mean of groups, excluding missing values.
df[['type1', 'type2', 'attack', 'defense']].groupby(['type1', 'type2']).mean()
```

```python
# Compute count of group, excluding missing values.
df[['type1', 'type2', 'attack', 'defense']].groupby(['type1', 'type2']).count()
```

```python
# all methods/attributes of DataFrameGroupBy class
[elem for elem in dir(pd.core.groupby.DataFrameGroupBy) if '_' not in elem]
```
